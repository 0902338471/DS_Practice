```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```
lgLCKSKT=lg[(lg['League']=='LCK')&(lg['Season']=='Spring')&(lg['Type']=='Season')&((lg['blueTeamTag']=='SKT')|(lg['redTeamTag']=='SKT'))]
lgLCKSKT
```

```
#dropna row
lgLCKSKT=lgLCKSKT.dropna()

blueTeam=list(lgLCKSKT['blueTeamTag'].unique())

redTeam=list(lgLCKSKT['redTeamTag'].unique())

#get List Team Name

listTeamName=list(set(blueTeam + redTeam))

print(listTeamName)

```

```
ratePerYear=[]


lgLCKSKT.reset_index(drop=True)

yearlist=lgLCKSKT['Year'].unique()
for year in yearlist:
  #get total matches of team
  
  curdf=lgLCKSKT[lgLCKSKT['Year']==year]
  totalMatches=len(curdf)
  
  print('Team SKT has played '+str(totalMatches)+' matches in year '+str(year))
  
  #get team members when SKT is blue Team
  blueTeamSKT=curdf[curdf['blueTeamTag']=='SKT']
  
  teamMemberBlue=set(pd.concat([blueTeamSKT['blueMiddle'],blueTeamSKT['blueTop'],blueTeamSKT['blueSupport'],blueTeamSKT['blueJungle'],blueTeamSKT['blueADC']]).unique())
  
  #print(teamMemberBlue)
   #get team members when SKT is redTeam
  redTeamSKT=curdf[curdf['redTeamTag']=='SKT']
  
  teamMemberRed=set(pd.concat([redTeamSKT['redMiddle'],redTeamSKT['redTop'],redTeamSKT['redSupport'],redTeamSKT['redJungle'],redTeamSKT['redADC']]).unique())
  
  #print(teamMemberRed)
  
  #join two team
  
  teamMember=teamMemberBlue
  
  teamMember.update(teamMemberRed)
  
  print('Team members this year:',teamMember)
  
  #get total winning matches
  totalWinningMatches=len(curdf[((curdf['blueTeamTag']=='SKT') & (curdf['bResult']==1))|((curdf['redTeamTag']=='SKT')&(curdf['rResult']==1))])
  
  print('Team SKT has won '+str(totalWinningMatches)+' matches in year '+str(year))
  
  ratePerYear.append((totalWinningMatches/totalMatches)*100)
```

```
sktRatePerYear=pd.DataFrame({'Year':yearlist,'WinningRate':ratePerYear})
sktRatePerYear.plot(x='Year',y='WinningRate',kind='bar')
```

```
f, axes = plt.subplots(2, 2, figsize=(12, 6))
i=0
j=0
for year in yearlist:
  lgYear=lg[lg["Year"]==year]
  
  lgFakerBlue=lgYear[lgYear["blueMiddle"]=="Faker"]["blueMiddleChamp"].reset_index()
  
  lgFakerBlue.rename(columns={'blueMiddleChamp':'Champ'}, 
                 inplace=True)
  
  lgFakerRed=lgYear[lgYear["redMiddle"]=="Faker"]["redMiddleChamp"].reset_index()
  
  lgFakerRed.rename(columns={'redMiddleChamp':'Champ'}, 
                 inplace=True)
  
  
  lgFakerChamp=lgFakerBlue.append(lgFakerRed,ignore_index = True)['Champ']
  
  lgFakertop10=lgFakerChamp.value_counts().head(10)

  print(lgFakertop10)

  lgFakertop10.plot(ax=axes[i][j],kind='barh')
  
  axes[i, j].set_title('Year '+str(year))
  
  j+=1
  if(j>1):
    i+=1
    j=0
```

```
#top 10 Champ being banned through year 
f1, axes1 = plt.subplots(3, 2, figsize=(20, 12))
i=0
j=0
yearlist=lg['Year'].unique()
for year in yearlist:
  lgYear=lg[lg["Year"]==year]
  
  lgChampBans=pd.Series()
  tmp=pd.concat([lgYear['blueBans'],lgYear['redBans']])
  
  tmp.rename('ChampBans')
  
  #print(tmp)
  
  for index, value in tmp.items():
    value=value[1:-1]
    value=value.split(', ')
    #print(pd.Series(value))
    lgChampBans=lgChampBans.append(pd.Series(value),ignore_index = True)
  
  #print(lgChampBans)
  
  #lgFakerChamp=lgFakerBlue.append(lgFakerRed,ignore_index = True)['Champ']
  
  lgtop10ChampBans=lgChampBans.value_counts().head(10)
  
  print(lgtop10ChampBans)

  lgtop10ChampBans.plot(ax=axes1[i][j],kind='barh')
  
  
  axes1[i, j].set_title('Year '+str(year))
  
  j+=1
  if(j>1):
    i+=1
    j=0
```
